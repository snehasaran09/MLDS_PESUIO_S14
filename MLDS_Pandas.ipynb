## Pandas Tutorial 

This notebook contains basic operations in Pandas and the approach to a problem. It also contains the explanation of the parameters. 

```python
import pandas as pd 
import numpy as np 
#pandas is a framework built on top of numpy .
```

Importing a Dataset 

```python
data = pd.read_csv('Salary_Data.csv') #function to read a csv file
```

Displaying the Loaded data

```python
data.head() #head() displays the first 5 rows in the dataset head(n) will display the first n values
```

```python
data.tail() #displays the last 5 rows of the dataset. tails(n) will display the last n rows of the dataset
```

```python
data.index #the number of rows of the dataset are displayed
```

```python
data.columns # displays the attributes(features) in the dataset along with their datatype
```

```python
data.describe() #displays a statistics summary of the dataset .
#The 25%, 50% and 75% are quartiles that split sorted data into 4 parts. 
```

```python
data.T #transposing your dataset
```

```python
data.sort_values(by='YearsExperience', ascending=False) 
#sorts the values in the dataset by the feature mentioned in (by='feature')
#by default sorts in ascending but ascending =False gives the descending order 
```

```python
data.sort_index(axis=1,ascending = True) # sorts the dataset by index labels 
```

A dataframe in Pandas(a 2-D data structure more like an array) can be converted to a numpy representation. However this might be a tedious process if the elements are of diff datatypes. 
The fundamental difference between numpy and pandas is that the array in numpy has to contain data of the same type whereas the pandas dataframe can contain objects of any datatype. 

```python
data.to_numpy()
```

Selecting Values

```python
data['Salary'].head() #displays the values of a single column 
```

```python
data[7:15]
 #displays the values from the row numbers specified, in this case from 7th till 15th row
```

```python
data.iloc[3:5] #selects a specific row or column from the dataset
```

Dealing with missing data

```python
data_new = pd.read_csv('creditcard.csv') #reading a new dataset with missing values
```

```python
data_new.shape
```

```python
data_new.isnull() #checks for null values 
```

```python
data_new.isnull().sum() #sum of null values
```

```python
df_new = data_new.copy() 
df_new.dropna() #dropping the null values from the dataset 
```

```python
data_new.fillna(value = data_new.mean()) 
#it can be seen that the null values have been replaced with mean of the column values
```

```python
data_new.mean() # displays the mean of each column 
```

```python
data_new.median() #median of each feature 
```

```python
data_new.mode() #mode of each column 
```

Mean , median and mode together are known as measures of central tendency . 
Measures of central tendencies are measures that describe the central value of a dataset or a set of values in general.
