## **Seaborn Tutorial**

Seaborn is more comfortable working with the pandas dataframe as comapred to Matplotlib. 
Seaborn works with a dataset as a whole and is more intuitive generally

```python
import seaborn as sns
import numpy as np 
import pandas as pd
```

1. Plotting pair-wise data relationships 

```python
data = pd.read_csv('IRIS.csv')
data
```

a. Pair Grid - allows you to draw a grid of different subplots , wherein each of them show a different relationship . 

```python
p = sns.PairGrid(data,hue='species') # pairgrid allows you to draw a grid dof small subplots of diff types 
p.map_upper(sns.scatterplot) # the plots on the upper diagonal are scatterplots
p.map_lower(sns.kdeplot) #  the lower diagomal consist of density plots
p.map_diag(sns.histplot)  # the diagonal plots are histograms
p.add_legend()
```

b. Pairplot - a shortcut function that simplifies the pairGrid . It is a combination of histograms and scatterplot. 

```python
sns.pairplot(data,hue='species',palette='Set2',diag_kind='kde') 
#hue is the colour variation, diag_kind sets the kind of plot in the diagonal elements
```

c. Heatmap - helps you visualise matrix like data (relation between features)

```python
data_copy = data.copy()
del data_copy['species']
#heatmap cannot work with string data , hence we make a copy of the dataset and delete the column 
corr = data_copy.corr() #finding the correlation 
sns.heatmap(corr, cmap='YlGnBu') #cmap is used to alter the colour of the heatmap
```

D. Joint Distribution Plot - Combine the information from scatterplot and histograms. 

```python
sns.jointplot(x='sepal_length', y='petal_length',data=data )
```

e. Stacked Bar Graph - each bar is visually divided into sub bars to represent multiple columns at once

```python
import matplotlib.pyplot as plt
#this is done using matplotlib library 
```

```python
data.plot(kind='bar',stacked=True,color=['Red','Black','Green','Blue'])
plt.show()
```
