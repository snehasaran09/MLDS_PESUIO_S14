# **Matplotlib Tutorial**

```python
import matplotlib.pyplot as plt 
import numpy as np 
```

Basic Plots 

1. Line Graph 

```python
x = [6,2,5]
y = [1,3,8]
plt.plot(x,y) #plot function to plot x and y values
plt.title('Simple Line Graph') #title for the grpah 
plt.xlabel('X Values') #title for the x axis
plt.ylabel('Y Values') #title for the y axis
plt.show()
```

```python
x = np.random.randint(0,50,10,dtype = int) #0 - low limit 50 - upper limit 10- size
y = ['A','B','C','D','E','F','G','H','I','J']
x
```

2. Bar Graph an Scatter Plot 

```python
plt.figure(figsize=(30,30)) #optional , used to fix the size of the graph . If not specified default values are considered
plt.subplot(5,5,1) #used to design the layout of the plots (rows,cols,plot_no). It can also be used while plotting multiple plots together
plt.title('Bar Graph')
plt.bar(x,y, color='green') #bar graph
plt.subplot(5,5,2)
plt.title('Scatter Plot')
plt.scatter(x,y,color='Purple',s=60) #scatter plot s = the size of the markers
plt.subplot(5,5,3)
plt.title('Horizontal Bar Graph')
plt.barh(x,y) #plotting a horizontal bar graph 
```

3. Stem Plot

```python
plt.stem(x,y)
```

4. Pie Charts

```python
plt.figure(figsize=(10,10))
plt.pie(x,labels=y,autopct='%1.1f%%')
```

5. Boxplot

```python
import pandas as pd
df = pd.DataFrame(np.random.rand(10,5), columns=['A','B','C','D','E'])
plt.boxplot(df)
```
