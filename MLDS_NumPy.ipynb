Creating an array in NumPy - 
Since python does not support arrays as such (python has lists), numpy allows you to create arrays. Arrays must have homogenous data unlike python lists. 

```python
a1 = np.array([1,2,6,7]) # 1-Dimension Array
```

```python
a2 = np.array([[1,2,3,4],[7,8,9,10]]) 
#2-Dimension Array . The dimensions of an array are called axes. 
```

```python
a2.ndim  #to check the dimension of an array 
```

```python
dim_arr = np.array([
        [5, 3, 7, 1],
        [2, 6, 7 ,9],
        [1, 1, 1, 1],
        [4, 3, 2, 0],
    ])
```

```python
new_array = a1.reshape(2,2)
new_array 
#reshaping a 1-D array into 2 arrays with 2 elements each
```

Difference between Copy and View

```python
x = a1.copy()
x
```

```python
y = a1.view()
a1[3]=17
print(a1)
print(y)
print(x)
# copy is not affected by changes made in the original array 
#view is affected by the changes made in the original array. 
```

Creating an Array from existing Data

```python
new_a1 = np.array([[1,1],[2,2]])
new_a2 = np.array([[3,3],[4,4]])

#stacking two arrays vertically 
# PLEASE NOTE - while stacking the arrays , the dimensions of the arrays have to be same
stacked_a = np.vstack((new_a1,new_a2))
stacked_a
```

```python
#splitting the array into 2 equally shaped arrays 
np.hsplit(stacked_a,2)  #vsplit instead of hsplit gives you a horizontal partition of the arrays
```

Basic Array operations

Iterating through a N-D array 

```python
for x in np.nditer(stacked_a): 
  print(x)

#for iterating using for loops , for each element we might need n for loops which is a tedious process to write for high dimensionalities
#nditer() is a function that can be used to access elements from a 1-D array to an n-D array
```

```python
for index, i in np.ndenumerate(stacked_a):
  print(index,i)

#ndenumerate() function gives you the index of the element along with the element
```

```python
x = np.where(stacked_a==1)
x
#gives you the index locations of the elements if present , else gives empty array
```

```python
rand_arr = np.random.default_rng()
rand_arr.random(5)

#np.random.default_rng() a function to generate random numbers
#random(n) generates an array containing 5 random elements
```

```python
unique_val = np.unique(stacked_a)
unique_val
#fetches the unique values in the array 
```

```python
np.unique(stacked_a).sum()
#sum of the unique values 
```

```python
rev_arr = np.flip(stacked_a)
rev_arr
#reversing the array 
```
